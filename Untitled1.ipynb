```python
#  Step 1: Import Libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from collections import Counter

%matplotlib inline
sns.set(style='whitegrid')

#  Step 2: Upload File (run this cell and upload your file)
from google.colab import files
uploaded = files.upload()

#  Step 3: Load Dataset
df = pd.read_csv(next(iter(uploaded)))
df.head()

#  Step 4: Basic Info and Null Values
print(df.info())
print("\nMissing values:\n", df.isnull().sum())

#  Step 5: Data Cleaning
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')
df['director'] = df['director'].fillna('Unknown')
df['cast'] = df['cast'].fillna('Unknown')
df['country'] = df['country'].fillna('Unknown')
df['rating'] = df['rating'].fillna('Unknown')
df['date_added'] = df['date_added'].ffill()
df['duration'] = df['duration'].fillna('Unknown')


#  Step 6: Content Type Count
sns.countplot(data=df, x='type', palette='Set2')
plt.title('Content Type Distribution')
plt.show()

#  Step 7: Top 10 Countries
top_countries = df['country'].value_counts().head(10)
sns.barplot(x=top_countries.values, y=top_countries.index, palette='magma')
plt.title('Top 10 Content-Producing Countries')
plt.xlabel('Count')
plt.show()

#  Step 8: Release Year Distribution
plt.figure(figsize=(10,5))
sns.histplot(df['release_year'], bins=30, kde=False, color='skyblue')
plt.title('Number of Releases Over the Years')
plt.xlabel('Release Year')
plt.ylabel('Count')
plt.show()

#  Step 9: Top Genres
genres = ', '.join(df['listed_in']).split(', ')
genre_counts = Counter(genres)
genre_df = pd.DataFrame(genre_counts.most_common(10), columns=['Genre', 'Count'])

sns.barplot(data=genre_df, x='Count', y='Genre', palette='viridis')
plt.title('Top 10 Genres on Netflix')
plt.show()

#  Step 10: Movie Duration Distribution
movies = df[df['type'] == 'Movie'].copy()
movies['minutes'] = movies['duration'].str.extract('([0-9]+)').astype(float)

plt.figure(figsize=(10,5))
sns.histplot(movies['minutes'].dropna(), bins=30, kde=True, color='green')
plt.title('Distribution of Movie Durations')
plt.xlabel('Duration (minutes)')
plt.show()
```
